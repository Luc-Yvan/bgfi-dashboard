```python
# ═══════════════════════════════════════════════════════════════════════════
# CELLULE 0 — IMPORTS GLOBAUX + CHEMINS + FONCTIONS UTILITAIRES
# ⚠️ Exécuter EN PREMIER. Toutes les dépendances sont ici.
# ═══════════════════════════════════════════════════════════════════════════

import os
import base64
import warnings
import numpy as np
import pandas as pd
import matplotlib
matplotlib.use('Agg')   # OBLIGATOIRE pour les serveurs / nbconvert
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.patches as mpatches
import matplotlib.ticker as mticker
from IPython.display import display, HTML, Image

warnings.filterwarnings('ignore')

# ── Paramètres globaux ────────────────────────────────────────────────────
TARGET_BANK  = 'BGFI'
ANNEE_DEBUT  = 2015
ANNEE_FIN    = 2020

# ── Palette BGFI Bank officielle ─────────────────────────────────────────
C_BLUE       = '#004a99'   # Bleu BGFI institutionnel (primaire)
C_GOLD       = '#E8A020'   # Or BGFI (couleur secondaire officielle)
C_GOLD_LIGHT = '#F5CF7E'   # Or clair / dégradé léger
C_BLUE_LIGHT = '#6699CC'   # Bleu clair pour barres secondaires
C_BGFI_HI   = '#C9A227'   # Surbrillance BGFI = or
C_AUTRE      = '#B0C4DE'   # Bleu pâle pour autres banques (contraste)
C_GRIS       = '#808080'   # Gris source
C_BANDEAU    = '#003087'   # Bleu foncé bandeau titre

# ── Chemins ────────────────────────────────────────────────────────────────
# ⚠️ NE PAS MODIFIER — Chemins validés en local
path_data = r"C:\Users\user\Documents\Data Ingenieur2\Projet_banque\Data\BASE_SENEGAL2.xlsx"
path_logo = r"C:\Users\user\Documents\Data Ingenieur2\Projet_banque\Data\logo_bgfi.jfif"

# ── Fonctions utilitaires ──────────────────────────────────────────────────

def calcul_tcam(v0, v1, n):
    """Taux de Croissance Annuel Moyen."""
    try:
        if v0 <= 0 or pd.isna(v0) or pd.isna(v1) or n <= 0:
            return 0.0
        return round(((v1 / v0) ** (1 / n) - 1) * 100, 1)
    except:
        return 0.0

def style_ax(ax):
    """Style épuré commun à tous les graphiques."""
    ax.spines[['top', 'right']].set_visible(False)
    ax.spines['left'].set_color('#dddddd')
    ax.spines['bottom'].set_color('#dddddd')
    ax.tick_params(colors='#444444', length=3)
    ax.set_facecolor('white')

# ── LOGO : placement cohérent sur TOUTES les pages ─────────────────────────
# Le logo est placé en BAS À DROITE, HORS de la zone graphique.
# La figure est systématiquement ajustée pour laisser de la place (rect).
LOGO_AX_RECT  = [0.830, 0.015, 0.155, 0.085]  # bas-droite, sous la zone graphique
LAYOUT_RECT   = [0,    0.13, 1.0,   0.94]    # laisse 13% en bas pour logo + source

def ajouter_logo(fig):
    """Incruste le logo BGFI en bas à droite, hors des graphiques, taille fixe."""
    try:
        logo = mpimg.imread(path_logo)
        ax_logo = fig.add_axes(LOGO_AX_RECT, anchor='SE', zorder=10)
        ax_logo.imshow(logo)
        ax_logo.axis('off')
    except Exception as e:
        print(f"⚠️ Logo non chargé : {e}")

def get_logo_base64():
    """Encode le logo en base64 pour la page de garde HTML."""
    try:
        with open(path_logo, 'rb') as f:
            return base64.b64encode(f.read()).decode('utf-8')
    except:
        return ''

def sauvegarder_afficher(fig, nom_fichier):
    """Sauvegarde la figure en PNG et l'affiche dans le HTML généré."""
    plt.savefig(nom_fichier, dpi=120, bbox_inches='tight')
    plt.close(fig)
    display(Image(filename=nom_fichier))

```

```python
# ═══════════════════════════════════════════════════════════════════════════
# CELLULE 3 — TOTAL BILAN DES BANQUES 2020
# Couleurs BGFI : barres bleues, BGFI en or
# Logo en bas à droite, hors graphique
# ═══════════════════════════════════════════════════════════════════════════

d_plot     = df_2020.sort_values('BILAN', ascending=True)
nb_banques = len(d_plot)
total_mds  = round(total_bilan_2020 / 1_000, 0)
couleurs   = [C_GOLD if s == TARGET_BANK else C_BLUE for s in d_plot['Sigle']]

fig, ax = plt.subplots(figsize=(14, 9), facecolor='white')
ax.set_facecolor('white')

bars = ax.barh(d_plot['Sigle'], d_plot['BILAN'], color=couleurs, height=0.7)

# Valeurs au bout des barres
max_val = d_plot['BILAN'].max()
for bar in bars:
    width = bar.get_width()
    ax.text(width + (max_val * 0.01), bar.get_y() + bar.get_height() / 2,
            f'{width:,.0f}', va='center', fontsize=9, color='#333333')

# BGFI en gras / or
for label in ax.get_yticklabels():
    if label.get_text() == TARGET_BANK:
        label.set_fontweight('bold')
        label.set_color(C_GOLD)

plt.suptitle(
    f'Le secteur Bancaire Sénégalais compte {nb_banques} banques représentant un\n'
    f'total bilan de {total_mds:,.0f} milliards de FCFA soit près de 60% du PIB national.',
    fontsize=14, fontweight='bold', x=0.5, y=0.97, color=C_BLUE
)
ax.set_title(f'Total Bilan des Banques, {ANNEE_FIN} (millions FCFA)',
             fontsize=11, style='italic', color='#555555', pad=12)

ax.spines[['top', 'right', 'bottom']].set_visible(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='y', length=0)

# Légende couleurs
leg = [mpatches.Patch(color=C_BLUE, label='Autres banques'),
       mpatches.Patch(color=C_GOLD, label=f'{TARGET_BANK} Bank')]
ax.legend(handles=leg, loc='lower right', fontsize=9, frameon=False)

# Ligne dorée source
fig.text(0.01, 0.09, 'Source: Commission Bancaire, APBEF-Sénégal,Power By Luc-Yvan',
         fontsize=9, color=C_GRIS, style='italic')

plt.tight_layout(rect=LAYOUT_RECT)
ajouter_logo(fig)
sauvegarder_afficher(fig, 'temp_graph_1.png')
```

## 1-2 Un Marché en Croissance — Évolution 2015–2020

```python
# ═══════════════════════════════════════════════════════════════════════════
# CELLULE 4 — ÉVOLUTION DU MARCHÉ (4 indicateurs)
# Couleurs BGFI : bleu + or pour TCAM, source en bas
# ═══════════════════════════════════════════════════════════════════════════

evolution_marche = df.groupby('ANNEE')[['BILAN', 'EMPLOI', 'RESSOURCES', 'COMPTE']].sum().reset_index()

metrics = [
    ('BILAN',      'Total Bilan',          C_BLUE),
    ('EMPLOI',     'Emplois (Crédits)',     C_BLUE),
    ('RESSOURCES', 'Ressources (Dépôts)',   C_BLUE),
    ('COMPTE',     'Nombre de Comptes',     C_BLUE),
]

fig, axs = plt.subplots(2, 2, figsize=(14, 8), facecolor='white')

for i, (col, title, color) in enumerate(metrics):
    ax   = axs.flat[i]
    bars = ax.bar(evolution_marche['ANNEE'].astype(str),
                  evolution_marche[col], color=color, width=0.6)

    v0 = evolution_marche[evolution_marche['ANNEE'] == ANNEE_DEBUT][col].values[0]
    v1 = evolution_marche[evolution_marche['ANNEE'] == ANNEE_FIN][col].values[0]
    t  = calcul_tcam(v0, v1, ANNEE_FIN - ANNEE_DEBUT)

    for bar in bars:
        h = bar.get_height()
        ax.text(bar.get_x() + bar.get_width() / 2., h + h * 0.01,
                f'{h:,.0f}', ha='center', va='bottom', fontsize=8, color=C_BLUE)

    # Annotation TCAM en or
    ax.set_title(f'{title}', fontweight='bold', color=C_BLUE, fontsize=10)
    ax.text(0.98, 0.96, f'TCAM : +{t}%',
            transform=ax.transAxes, ha='right', va='top',
            fontsize=10, fontweight='bold', color=C_GOLD,
            bbox=dict(boxstyle='round,pad=0.3', facecolor='white',
                      edgecolor=C_GOLD, linewidth=1.5))

    ax.spines[['top', 'right']].set_visible(False)
    ax.spines[['left','bottom']].set_color('#dddddd')
    ax.get_yaxis().set_visible(False)
    ax.tick_params(axis='x', labelsize=9)

fig.text(0.5, 0.97,
    'Un marché bancaire de taille moyenne, compétitif et en croissance soutenue.',
    ha='center', fontsize=13, fontweight='bold', color=C_BLUE)

fig.text(0.01, 0.09, 'Source: Commission Bancaire, APBEF-Sénégal,Power By Luc-Yvan',
         fontsize=9, color=C_GRIS, style='italic')

plt.tight_layout(rect=LAYOUT_RECT)
ajouter_logo(fig)
sauvegarder_afficher(fig, 'temp_graph_2.png')
```

## 1-4 Positionnement Stratégique — BGFI vs Toutes les Banques

```python
# ═══════════════════════════════════════════════════════════════════════════
# CELLULE 6 — POSITIONNEMENT BGFI vs TOUTES LES BANQUES
# BGFI = point or (C_GOLD), autres = bleu clair
# ═══════════════════════════════════════════════════════════════════════════

metrics   = ['BILAN', 'EMPLOI', 'RESSOURCES', 'COMPTE']
titles    = ['Total Bilan', 'Emplois (Crédits)', 'Ressources (Dépôts)', 'Nombre de Comptes']
all_banks = df['Sigle'].unique()

fig, axs = plt.subplots(2, 2, figsize=(16, 12), facecolor='white')

for i, m in enumerate(metrics):
    ax           = axs.flat[i]
    total_m_2020 = df_2020[m].sum()

    for b in all_banks:
        df_b    = df[df['Sigle'] == b].sort_values('ANNEE')
        v_start = df_b[df_b['ANNEE'] == ANNEE_DEBUT][m].values
        v_end   = df_b[df_b['ANNEE'] == ANNEE_FIN][m].values

        if len(v_start) > 0 and len(v_end) > 0:
            ms      = (v_end[0] / total_m_2020) * 100
            growth  = calcul_tcam(v_start[0], v_end[0], 5)
            is_bgfi = (str(b).upper() == TARGET_BANK)

            ax.scatter(ms, growth,
                       color=C_GOLD if is_bgfi else C_BLUE,
                       s=350 if is_bgfi else 100,
                       edgecolor='white', linewidth=1.2,
                       alpha=1.0 if is_bgfi else 0.7,
                       zorder=5 if is_bgfi else 3)

            ax.annotate(f'{b}', (ms, growth),
                        xytext=(5, 5), textcoords='offset points',
                        fontsize=8,
                        fontweight='bold' if is_bgfi else 'normal',
                        color=C_GOLD if is_bgfi else '#666666')

    ax.axhline(0, color='#cccccc', linewidth=0.8, linestyle='--')
    ax.set_title(f'Positionnement : {titles[i]}', fontweight='bold',
                 fontsize=11, color=C_BLUE)
    ax.set_xlabel('Part de Marché 2020 (%)', fontsize=9, color=C_GRIS)
    ax.set_ylabel('TCAM 2015-2020 (%)', fontsize=9, color=C_GRIS)
    ax.spines[['top', 'right']].set_visible(False)
    ax.spines[['left','bottom']].set_color('#dddddd')
    ax.grid(True, linestyle=':', alpha=0.3)

# Légende commune
leg = [plt.scatter([], [], c=C_GOLD, s=120, label=f'{TARGET_BANK} Bank'),
       plt.scatter([], [], c=C_BLUE, s=60,  label='Autres banques', alpha=0.7)]
fig.legend(handles=leg, loc='lower center', ncol=2, fontsize=10,
           frameon=False, bbox_to_anchor=(0.5, 0.0))

plt.suptitle(
    f'Cartographie Concurrentielle : {TARGET_BANK} Bank vs Secteur Bancaire Sénégalais',
    fontsize=15, fontweight='bold', color=C_BLUE, y=0.98
)

fig.text(0.01, 0.09, 'Source: Commission Bancaire, APBEF-Sénégal,Power By Luc-Yvan',
         fontsize=9, color=C_GRIS, style='italic')

plt.tight_layout(rect=LAYOUT_RECT)
ajouter_logo(fig)
sauvegarder_afficher(fig, 'pos_complet_bgfi_bleu.png')
```

## 1-5 Positionnement des Groupes Bancaires

```python
# ═══════════════════════════════════════════════════════════════════════════
# CELLULE 7 — POSITIONNEMENT DES GROUPES BANCAIRES
# ═══════════════════════════════════════════════════════════════════════════

df_gp = df.groupby(['Goupe_Bancaire', 'ANNEE'])[['BILAN', 'EMPLOI', 'RESSOURCES', 'COMPTE']].sum().reset_index()

def scatter_groupes(ax, titre, metric):
    total_m_2020 = df_gp[df_gp['ANNEE'] == ANNEE_FIN][metric].sum()
    groupes_uniq = df_gp['Goupe_Bancaire'].unique()

    for gp in groupes_uniq:
        d_g     = df_gp[df_gp['Goupe_Bancaire'] == gp].sort_values('ANNEE')
        v_start = d_g[d_g['ANNEE'] == ANNEE_DEBUT][metric].values
        v_end   = d_g[d_g['ANNEE'] == ANNEE_FIN][metric].values

        if len(v_start) > 0 and len(v_end) > 0:
            ms         = (v_end[0] / total_m_2020) * 100
            growth     = calcul_tcam(v_start[0], v_end[0], 5)
            is_bgfi_gp = (gp == GROUPE_BGFI)

            ax.scatter(ms, growth,
                       color=C_GOLD if is_bgfi_gp else C_BLUE,
                       s=500 if is_bgfi_gp else 200,
                       edgecolor='white', linewidth=1.5,
                       zorder=5 if is_bgfi_gp else 3)

            if ms > 1 or is_bgfi_gp:
                ax.annotate(f' {gp}', (ms, growth),
                            fontsize=8.5,
                            fontweight='bold' if is_bgfi_gp else 'normal',
                            color=C_GOLD if is_bgfi_gp else '#444444',
                            xytext=(5, 5), textcoords='offset points')

    ax.set_title(titre, fontsize=10, fontweight='bold', color=C_BLUE, pad=8)
    ax.set_xlabel('Part de Marché du Groupe (%)', fontsize=8, color=C_GRIS)
    ax.set_ylabel('TCAM (%)', fontsize=8, color=C_GRIS)
    ax.axhline(0, color='#cccccc', linestyle='--', linewidth=0.8)
    ax.spines[['top', 'right']].set_visible(False)
    ax.spines[['left','bottom']].set_color('#dddddd')
    ax.grid(True, linestyle=':', alpha=0.3)
    ax.set_facecolor('white')

fig, axs = plt.subplots(2, 2, figsize=(15, 11), facecolor='white')

scatter_groupes(axs[0, 0], 'Sur le nombre de comptes', 'COMPTE')
scatter_groupes(axs[0, 1], 'Sur le total bilan',       'BILAN')
scatter_groupes(axs[1, 0], 'Sur les ressources',       'RESSOURCES')
scatter_groupes(axs[1, 1], 'Sur les emplois',          'EMPLOI')

fig.text(0.01, 0.98,
    f'Positionnement des groupes bancaires — Marché Sénégalais',
    fontsize=14, fontweight='bold', va='top', color=C_BLUE)

# Bandeau bleu foncé
fig.text(0.5, 0.93,
    f'Le {GROUPE_BGFI} présente la dynamique de croissance la plus forte',
    ha='center', fontsize=10, fontweight='bold', color='white',
    bbox=dict(boxstyle='square,pad=0.5', facecolor=C_BANDEAU, edgecolor=C_GOLD, linewidth=1.5))

fig.text(0.01, 0.09, 'Source: Commission Bancaire, APBEF-Sénégal,Power By Luc-Yvan',
         fontsize=9, color=C_GRIS, style='italic')

plt.tight_layout(rect=LAYOUT_RECT)
ajouter_logo(fig)
sauvegarder_afficher(fig, 'pos_groupes_original.png')
```

## 1-6 Positionnement de BGFI au sein de son Groupe

```python
# ═══════════════════════════════════════════════════════════════════════════
# CELLULE 8 — BGFI vs LES BANQUES DE SON GROUPE
# ═══════════════════════════════════════════════════════════════════════════

df_groupe      = df_2020[df_2020['Goupe_Bancaire'] == GROUPE_BGFI].copy()
banques_groupe = df_groupe['Sigle'].unique()

def scatter_bnk(ax, titre, metric):
    total_m = df_2020[metric].sum()

    for b in banques_groupe:
        df_b    = df[df['Sigle'] == b].sort_values('ANNEE')
        v_start = df_b[df_b['ANNEE'] == ANNEE_DEBUT][metric].values
        v_end   = df_b[df_b['ANNEE'] == ANNEE_FIN][metric].values

        if len(v_start) > 0 and len(v_end) > 0:
            ms      = (v_end[0] / total_m) * 100
            growth  = calcul_tcam(v_start[0], v_end[0], 5)
            is_bgfi = (b == TARGET_BANK)

            ax.scatter(ms, growth,
                       color=C_GOLD if is_bgfi else C_BLUE,
                       s=400 if is_bgfi else 140,
                       edgecolor='white', linewidth=1.5,
                       zorder=5 if is_bgfi else 3)

            ax.annotate(f' {b}', (ms, growth),
                        fontsize=9,
                        fontweight='bold' if is_bgfi else 'normal',
                        color=C_GOLD if is_bgfi else '#555555',
                        xytext=(5, 5), textcoords='offset points')

    ax.set_title(titre, fontsize=10, fontweight='bold', color=C_BLUE, pad=8)
    ax.set_xlabel('Part de Marché (%)', fontsize=8, color=C_GRIS)
    ax.set_ylabel('TCAM (%)', fontsize=8, color=C_GRIS)
    ax.axhline(0, color='#cccccc', linestyle='--', linewidth=0.8)
    ax.grid(True, linestyle=':', alpha=0.3)
    ax.spines[['top', 'right']].set_visible(False)
    ax.spines[['left','bottom']].set_color('#dddddd')
    ax.set_facecolor('white')

fig, axs = plt.subplots(2, 2, figsize=(15, 11), facecolor='white')

scatter_bnk(axs[0, 0], 'Sur le nombre de comptes', 'COMPTE')
scatter_bnk(axs[0, 1], 'Sur le total bilan',       'BILAN')
scatter_bnk(axs[1, 0], 'Sur les ressources',       'RESSOURCES')
scatter_bnk(axs[1, 1], 'Sur les emplois',          'EMPLOI')

fig.text(0.01, 0.98,
    f'{TARGET_BANK} a une dynamique plus forte sur tous les indicateurs.',
    fontsize=13, fontweight='bold', va='top', color=C_BLUE)

fig.text(0.5, 0.93,
    f'Positionnement de {TARGET_BANK} par rapport aux banques du {GROUPE_BGFI}',
    ha='center', fontsize=10, fontweight='bold', color='white',
    bbox=dict(boxstyle='square,pad=0.5', facecolor=C_BANDEAU, edgecolor=C_GOLD, linewidth=1.5))

fig.text(0.01, 0.09, 'Source: Commission Bancaire, APBEF-Sénégal,Power By Luc-Yvan',
         fontsize=9, color=C_GRIS, style='italic')

plt.tight_layout(rect=LAYOUT_RECT)
ajouter_logo(fig)
sauvegarder_afficher(fig, 'pos_groupe_original.png')
```

## 1-7 Productivité Commerciale — BGFI vs Marché

```python
# ═══════════════════════════════════════════════════════════════════════════
# CELLULE 9 — PRODUCTIVITÉ BGFI VS MARCHÉ
# ═══════════════════════════════════════════════════════════════════════════

row_bgfi = df_2020[df_2020['Sigle'] == TARGET_BANK].iloc[0]

moy_empl_ag  = (df_2020['EMPLOI']     / df_2020['AGENCE']).replace([np.inf, -np.inf], np.nan).dropna().mean()
moy_ress_ag  = (df_2020['RESSOURCES'] / df_2020['AGENCE']).replace([np.inf, -np.inf], np.nan).dropna().mean()
moy_empl_eff = (df_2020['EMPLOI']     / df_2020['EFFECTIF']).replace([np.inf, -np.inf], np.nan).dropna().mean()
moy_ress_eff = (df_2020['RESSOURCES'] / df_2020['EFFECTIF']).replace([np.inf, -np.inf], np.nan).dropna().mean()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), facecolor='white')

x     = np.arange(2)
width = 0.35

# Graphique A : Par Agence
bgfi_ag   = [row_bgfi['EMPLOI'] / row_bgfi['AGENCE'], row_bgfi['RESSOURCES'] / row_bgfi['AGENCE']]
marche_ag = [moy_empl_ag, moy_ress_ag]

ax1.bar(x - width/2, bgfi_ag,   width, label=f'{TARGET_BANK} Bank', color=C_GOLD)
ax1.bar(x + width/2, marche_ag, width, label='Moyenne Marché',      color=C_BLUE)
ax1.set_title('Productivité par Agence (M FCFA)', fontweight='bold', color=C_BLUE)
ax1.set_xticks(x)
ax1.set_xticklabels(['Emplois / Agence', 'Ressources / Agence'])
ax1.legend(fontsize=9, frameon=False)
style_ax(ax1)

# Graphique B : Par Agent
bgfi_eff   = [row_bgfi['EMPLOI'] / row_bgfi['EFFECTIF'], row_bgfi['RESSOURCES'] / row_bgfi['EFFECTIF']]
marche_eff = [moy_empl_eff, moy_ress_eff]

ax2.bar(x - width/2, bgfi_eff,   width, label=f'{TARGET_BANK} Bank', color=C_GOLD)
ax2.bar(x + width/2, marche_eff, width, label='Moyenne Marché',      color=C_BLUE)
ax2.set_title('Productivité par Agent (M FCFA)', fontweight='bold', color=C_BLUE)
ax2.set_xticks(x)
ax2.set_xticklabels(['Emplois / Agent', 'Ressources / Agent'])
ax2.legend(fontsize=9, frameon=False)
style_ax(ax2)

plt.suptitle(f'Analyse de la Productivité : {TARGET_BANK} Bank vs Moyenne du Secteur',
             fontsize=14, fontweight='bold', color=C_BLUE, y=0.97)

fig.text(0.01, 0.09, 'Source: Commission Bancaire, APBEF-Sénégal,Power By Luc-Yvan',
         fontsize=9, color=C_GRIS, style='italic')

plt.tight_layout(rect=LAYOUT_RECT)
ajouter_logo(fig)
sauvegarder_afficher(fig, 'temp_graph_4.png')
```

## 1-8 Ranking Productivité — Toutes les Banques

```python
# ═══════════════════════════════════════════════════════════════════════════
# CELLULE 10 — RANKING COMPLET PRODUCTIVITÉ
# BGFI en or sur fond bleu, médiane en or
# ═══════════════════════════════════════════════════════════════════════════

df_prod = df_2020.copy()
df_prod['EMP_AG'] = df_prod['EMPLOI']     / df_prod['AGENCE']
df_prod['RES_AG'] = df_prod['RESSOURCES'] / df_prod['AGENCE']
df_prod = df_prod.replace([np.inf, -np.inf], 0)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 10), facecolor='white')

for ax, col, titre in [
    (ax1, 'EMP_AG', 'Emplois / Agence (M FCFA)'),
    (ax2, 'RES_AG', 'Ressources / Agence (M FCFA)'),
]:
    d      = df_prod.sort_values(col, ascending=True)
    colors = [C_GOLD if str(s).upper() == TARGET_BANK else C_BLUE for s in d['Sigle']]
    bars   = ax.barh(d['Sigle'], d[col], color=colors, height=0.7)

    # Ligne médiane or
    mediane = d[col].median()
    ax.axvline(mediane, color=C_GOLD, linewidth=2, linestyle='--', label=f'Médiane : {mediane:,.0f}')

    for bar in bars:
        ax.text(bar.get_width() + d[col].max()*0.01, bar.get_y() + bar.get_height() / 2,
                f'{bar.get_width():,.0f}', va='center', fontsize=8, color='#444444')

    ax.set_title(titre, fontweight='bold', color=C_BLUE, pad=12)
    ax.legend(fontsize=8, frameon=False)
    ax.spines[['top', 'right', 'bottom']].set_visible(False)
    ax.get_xaxis().set_visible(False)
    ax.tick_params(axis='y', length=0, labelsize=9)
    ax.set_facecolor('white')

    for label in ax.get_yticklabels():
        if TARGET_BANK in label.get_text().upper():
            label.set_fontweight('bold')
            label.set_color(C_GOLD)

plt.suptitle(
    f'Efficacité du Réseau Physique — Classement Intégral du Marché',
    fontsize=15, fontweight='bold', color=C_BLUE, y=0.98
)

fig.text(0.01, 0.09, 'Source: Commission Bancaire, APBEF-Sénégal,Power By Luc-Yvan',
         fontsize=9, color=C_GRIS, style='italic')

plt.tight_layout(rect=LAYOUT_RECT)
ajouter_logo(fig)
sauvegarder_afficher(fig, 'prod_all_banks_bgfi.png')
```

## 1-9 Solidité et Maîtrise des Risques

```python
# ═══════════════════════════════════════════════════════════════════════════
# CELLULE 11 — RISQUE & LEVIER FINANCIER
# ═══════════════════════════════════════════════════════════════════════════

df_risk = df_2020.copy()
df_risk['RISQUE_RATIO'] = (df_risk['COÛT.DU.RISQUE'] / df_risk['BILAN']) * 100
df_risk['LEVIER']       = df_risk['BILAN'] / df_risk['FONDS.PROPRE']
df_risk = df_risk.replace([np.inf, -np.inf], 0).fillna(0)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 11), facecolor='white')

for ax, col, titre, fmt in [
    (ax1, 'RISQUE_RATIO', 'Coût Net du Risque / Bilan (%)',       '{:.2f}%'),
    (ax2, 'LEVIER',       'Levier Financier (Bilan / Fonds Propres)', '{:.1f}x'),
]:
    d      = df_risk.sort_values(col, ascending=True)
    colors = [C_GOLD if str(s).upper() == TARGET_BANK else C_BLUE for s in d['Sigle']]
    bars   = ax.barh(d['Sigle'], d[col], color=colors, height=0.7)

    # Médiane or
    mediane = d[col].median()
    ax.axvline(mediane, color=C_GOLD, linewidth=2, linestyle='--', label=f'Médiane : {fmt.format(mediane)}')

    for bar in bars:
        ax.text(bar.get_width() + d[col].max()*0.01,
                bar.get_y() + bar.get_height() / 2,
                fmt.format(bar.get_width()), va='center', fontsize=8)

    ax.set_title(titre, fontweight='bold', color=C_BLUE, pad=12)
    ax.legend(fontsize=8, frameon=False)
    ax.spines[['top', 'right', 'bottom']].set_visible(False)
    ax.get_xaxis().set_visible(False)
    ax.tick_params(axis='y', length=0, labelsize=9)
    ax.set_facecolor('white')

    for label in ax.get_yticklabels():
        if TARGET_BANK in label.get_text().upper():
            label.set_fontweight('bold')
            label.set_color(C_GOLD)
            label.set_fontsize(10)

plt.suptitle(
    f'Solidité et Maîtrise des Risques — Classement Complet du Secteur',
    fontsize=15, fontweight='bold', color=C_BLUE, y=0.98
)

fig.text(0.01, 0.09, 'Source: Commission Bancaire, APBEF-Sénégal,Power By Luc-Yvan',
         fontsize=9, color=C_GRIS, style='italic')

plt.tight_layout(rect=LAYOUT_RECT)
ajouter_logo(fig)
sauvegarder_afficher(fig, 'risk_all_banks_bgfi.png')
```
